```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function of power-of-two and sum values 
def pow_two_func(pow2_n):
    pow2_array = []
    result = 0
    count = 0
    while count < pow2_n:
        pow2_array.append(2**count)
        result = result + 2**count
        count += 1
    result_fin = 2**pow2_n-1
#     print("So members moi the he: ", pow2_array)
    print("Tong so members: ",f"{result_fin:,}")
    return pow2_array, result_fin

def plot_scatter_func(pow2_array):
    fig, ax = plt.subplots(figsize=(24, 12), dpi=60)
    index_pow2 = []
    for i in range(1, len(pow2_array)+1):
        index_pow2.append(i)
        
    # set a title and labels
    ax.set_title('An application of Power of Two')
    ax.set_xlabel('Generation (g_n)')
    ax.set_ylabel('Number of your Family')

    plt.plot(range(len(index_pow2)), pow2_array, 'bo') # Plotting data
    plt.xticks(range(len(index_pow2)), index_pow2) # Redefining x-axis labels

    for i, v in enumerate(pow2_array):
        ax.annotate(str(v), xy=(i,v), xytext=(-7,7), textcoords='offset points')
    plt.show()

def my_program():
    while 3>1:
        print("Ban dang su dung chuong trinh ung dung power-of-two!")
        
        # tinh toan so the he theo nam khao sat
        so_nam = int(input("So nam khao sat, so_nam : "))
        
        gen_n = np.round(so_nam/30)
        print("So the he tao ra: ", gen_n)
        
        # function of power-of-two and sum values
        pow2_array, result_fin = pow_two_func(gen_n)
            
        print("#######################################################################")
        print("Sau", so_nam, "nam nua, ban se tao ra them: ", f"{result_fin:,}", "nguoi tren Trai Dat")
        print("#######################################################################")    
            
        # check for exist my_program
        keywork = input("Ban co muon plot ket qua, vui long say {yes}: ")
        if keywork == "yes":       
            print("Chuong trinh se plot chuoi power-of-two sau 3s ...")
            time.sleep(3)
            # ####################################################################
            # plot application
            # ####################################################################
            plot_scatter_func(pow2_array)
        
        # check for exist my_program
        keywork = input("Neu ban muon ket thuc chuong trinh, vui long say {bye}: ")
        if keywork == "bye":
            break
        else:
            print("########################################################")
            pass
```

```python
my_program()
```
