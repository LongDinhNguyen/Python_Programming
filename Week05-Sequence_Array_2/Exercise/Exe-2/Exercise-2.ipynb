```python
import pandas
import numpy as np
import time
from datetime import datetime as dt
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
sns.set_context("notebook", font_scale=1.3)
```

```python
# data_read = pandas.read_csv('x-ANACONDA/NLU-DH21HM/Week04-Sequence_Array/Exercise/IP04-exercise.csv')
data_read = pandas.read_csv('IP04-exercise.csv')
data_read.head()
# print(data_read.head())
data_read.info()
# data_read.set_index('NOC', inplace=True)
```

```python
countries = data_read[['NOC']]
# print(countries)
gold_cal = data_read[['GoldCal']]
# gold-cal
```

```python
countries = data_read.NOC.values
print(countries)
gold_cal = data_read.GoldCal.values
print(gold_cal)
```

```python
figure(figsize=(24, 10), dpi=80)
plt.title('So HCV cac quoc gia', fontsize=18, y=1.05)
plt.ylabel('So HCV', fontsize=18)
plt.xlabel('Quoc Gia', fontsize=18)
data_read_plot = gold_cal
plt.plot(countries, data_read_plot, 'ro-', linewidth=2, label='HCV')
plt.legend(['HCV'])
plt.show()
```

```python
# analysis 5 top countries in total 13 countries in term of gold medal
sum_medal = np.sum(gold_cal)
print(sum_medal)
```

```python
top_5 = []
for m in range(5):
    top_5.append(100*gold_cal[m]/sum_medal)
print(top_5)
```

```python
figure(figsize=(10, 8), dpi=80)
plt.title('top 5 / tong 13 quoc gia', fontsize=18, y=1.05)
mylabels = ["Norway", "Germany", "Canada", "United State", "Netherlands"]
data_read_plot = top_5
# mycolors = ["brown", "red", "blue", "green", "purple"]
mycolors = ["r", "b", "g", "c", "m"]
myexplode = [0.02, 0.02, 0.02, 0.02, 0.02]
# plt.pie(data_read_plot, labels = mylabels)
plt.pie(data_read_plot, labels = mylabels, startangle = 0, explode = myexplode, colors = mycolors)
# plt.pie(data_read_plot, labels = mylabels, startangle = 0, explode = myexplode, shadow = True)
plt.legend(title = "Top 5:")
plt.show()
```
